# Figure 8 plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,7))

plt.subplot(221)
res = np.load('results-Nf-1p6.npy',allow_pickle=True).item()
results = res['results']
v_i = res['v_i']
timesteps = res['timesteps']

plt.title(r'(a) $N_f=1.6$',fontsize=18,loc='left')
for i in range(v_i.size):
    h = results[i][0][int(0.5*timesteps.size):None]
    new_lens = results[i][1][int(0.5*timesteps.size):None]
    old_lens = np.roll(new_lens,-1)
    jumps = h[old_lens==1] - h[new_lens==1]
    plt.plot(v_i[i]*np.ones(h[old_lens==1].size),jumps,'o',markersize=2,color='k')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r'$z_\ell^\star - z_\mathrm{n} $',fontsize=18)
plt.xlabel(r'$v_\mathrm{i}$',fontsize=18) 
plt.xlim(0.068,0.074)
plt.ylim(3,8)


plt.subplot(222)
res = np.load('results-Nf-1p75.npy',allow_pickle=True).item()
results = res['results']
v_i = res['v_i']
timesteps = res['timesteps']

plt.title(r'(b) $N_f=1.75$',fontsize=18,loc='left')
for i in range(v_i.size):
    h = results[i][0][int(0.5*timesteps.size):None]
    new_lens = results[i][1][int(0.5*timesteps.size):None]
    old_lens = np.roll(new_lens,-1)
    jumps = h[old_lens==1] - h[new_lens==1]
    plt.plot(v_i[i]*np.ones(h[old_lens==1].size),jumps,'o',markersize=2,color='k')
plt.gca().yaxis.set_label_position("right")
plt.gca().yaxis.tick_right()    
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r'$z_\ell^\star - z_\mathrm{n} $',fontsize=18)
plt.xlabel(r'$v_\mathrm{i}$',fontsize=18) 
plt.xlim(0.054,0.06)
plt.ylim(3,8)


plt.subplot(223)
res = np.load('results-Nf-1p8.npy',allow_pickle=True).item()
results = res['results']
v_i = res['v_i']
timesteps = res['timesteps']

plt.title(r'(c) $N_f=1.8$',fontsize=18,loc='left')
for i in range(v_i.size):
    h = results[i][0][int(0.5*timesteps.size):None]
    new_lens = results[i][1][int(0.5*timesteps.size):None]
    old_lens = np.roll(new_lens,-1)
    jumps = h[old_lens==1] - h[new_lens==1]
    plt.plot(v_i[i]*np.ones(h[old_lens==1].size),jumps,'o',markersize=2,color='k')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r'$z_\ell^\star - z_\mathrm{n} $',fontsize=18)
plt.xlabel(r'$v_\mathrm{i}$',fontsize=18) 
plt.xlim(0.048,0.054)
plt.ylim(3,12)

plt.subplot(224)
res = np.load('results-Nf-2.npy',allow_pickle=True).item()
results = res['results']
v_i = res['v_i']
timesteps = res['timesteps']

plt.title(r'(d) $N_f=2$',fontsize=18,loc='left')
for i in range(v_i.size):
    h = results[i][0][int(0.5*timesteps.size):None]
    new_lens = results[i][1][int(0.5*timesteps.size):None]
    old_lens = np.roll(new_lens,-1)
    jumps = h[old_lens==1] - h[new_lens==1]
    plt.plot(v_i[i]*np.ones(h[old_lens==1].size),jumps,'o',markersize=2,color='k')
plt.gca().yaxis.set_label_position("right")
plt.gca().yaxis.tick_right()      
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r'$z_\ell^\star - z_\mathrm{n} $',fontsize=18)
plt.xlabel(r'$v_\mathrm{i}$',fontsize=18) 
plt.xlim(0.036,0.042)
plt.ylim(3,12)

plt.tight_layout()
plt.savefig('bifurcations')
plt.show()
plt.close()
```
